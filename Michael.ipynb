```python
from csv import reader
import re
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

fld = 'COVID-19/csse_covid_19_data/csse_covid_19_time_series/'
fldt = (fld+'/time_series_19-covid-%s.csv')

def is_monotonic(tt):
    return all(tt[i+1] - tt[i] >= 0 for i in range(len(tt)-1))

def l(n):
    ls = [re.split(r',(?=(?:"[^"]*?(?: [^"]*)*))|,(?=[^",]+(?:,|$))', l) 
          for l in open(fldt % n, 'r').read().strip().split('\n')]
    time_len = len(ls[0][4:])
    
    by_country = {}
    for c in ls[1:]:
        cc = c[1].strip()
        if cc not in by_country:
            by_country[cc] = [0 for _ in range(time_len)]
        for i in range(time_len):
            by_country[cc][i] += int(c[4+i])
    todel = []
    for k, v in by_country.items():
        if not is_monotonic(v):
            print('Not monotonic "', k, '"')
            todel.append(k)
    #for k in todel:
    #    del by_country[k]
    return by_country

def plot(tseries):
    sns.lineplot(data=[pd.Series(t) for t in tseries])

deaths = l('Deaths')
confirmed = l('Confirmed')
recovered = l('Recovered')
print(confirmed["China"])
#plot([pd.Series(deaths["China"]) / pd.Series(confirmed["China"])])
#plot([pd.Series(deaths["China"]) / pd.Series(recovered["China"])])

def showct(ct):
    plt.title(ct)
    plot([pd.Series(recovered[ct])])
    merge = pd.Series(deaths[ct]) + pd.Series(recovered[ct])
    plot([merge])
    plot([[0]*0+confirmed[ct]])
    plt.show()



def cl(ct, doplt=False, justrecovered=False):
    if justrecovered:
        merge = pd.Series(recovered[ct])
    else:
        merge = pd.Series(deaths[ct]) + pd.Series(recovered[ct])
    if doplt:
        pass
        #plot([merge])
        #plot([confirmed[ct]])
        #plt.show()
    
    vv = []
    for i in range(len(confirmed[ct])):
        t = confirmed[ct][i]
        if t <= 20:
            continue
        tj = next((j for j, v in enumerate(merge) if v >= t), None)
        if tj is None:
            continue
        vv.append(tj - i)
    if doplt:
        plt.title('%s lag' % ct)
        plot([vv])
        plt.show()
        
    return 0 if len(vv) == 0 else sum(vv) / len(vv), len(vv)

def dpct(ct, push=10):
    if len(deaths[ct]) <= push:
        return
    vv = []
    for i in range(push, len(deaths[ct])-1):
        dc = confirmed[ct][i-(push-1)] - confirmed[ct][i-push]
        dd = deaths[ct][i+1] - deaths[ct][i]
        print(dd, dc)
        if dc <= 5:
            continue
        vv.append(dd / dc)
    plt.title('%s Death Ratio' % ct)
    plot([vv])
    plt.show()

lags = {}
for k, v in confirmed.items():
    a, l = cl(k, justrecovered=False)
    if l >= 1:
        lags[k] = a

showct('Iran')
showct('China')
showct('Italy')
showct('Thailand')

cl('Iran', True)
cl('Italy', True)
cl('US', True)
cl('Germany', True)
cl('China', True)

dpct('Iran')
dpct('Italy')
dpct('US')
dpct('Germany')
dpct('China')

print(lags)
```
